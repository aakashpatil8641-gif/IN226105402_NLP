## Task 1: Replace Hardcoded Prompts


```python

task1_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

def explain_topic(topic: str) -> str:
    """Generate an explanation prompt for any topic using PromptTemplate."""
    return task1_template.format(topic=topic)

print("=== Task 1: Replace Hardcoded Prompts ===")
print(explain_topic("Machine Learning"))
print(explain_topic("Neural Networks"))
print(explain_topic("Python"))
```

## Task 2: Multi-Input Prompt System

```python
task2_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

def multi_input_prompt(topic: str, audience: str, tone: str) -> str:

    return task2_template.format(topic=topic, audience=audience, tone=tone)

test_cases = [
    ("AI",            "beginners", "friendly"),
    ("Python",        "kids",      "fun"),
    ("Deep Learning", "engineers", "technical"),
]

print("=== Task 2: Multi-Input Prompt System ===")
for topic, audience, tone in test_cases:
    print(multi_input_prompt(topic, audience, tone))
```

## Task 3: Prompt Variations Engine

```python
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

storytelling_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

VARIATION_TEMPLATES = {
    "teaching":     teaching_template,
    "interview":    interview_template,
    "storytelling": storytelling_template,
}

def prompt_variations_engine(topic: str) -> dict:
    return {
        style: template.format(topic=topic)
        for style, template in VARIATION_TEMPLATES.items()
    }

#  Run for the specified topic ──
topic = "Machine Learning"
variations = prompt_variations_engine(topic)

print("=== Task 3: Prompt Variations Engine ===")
for style, prompt in variations.items():
    print(f"[{style.upper()}] {prompt}")
```

## Task 4: ChatPromptTemplate System

```python
ROLE_SYSTEM_MESSAGES = {
    "teacher":     "You are an expert teacher. Explain concepts clearly and patiently.",
    "interviewer": "You are a technical interviewer. Ask probing, insightful questions.",
    "motivator":   "You are an enthusiastic motivator. Inspire and encourage the learner.",
}

def build_chat_prompt(role: str, topic: str) -> list:

    system_message = ROLE_SYSTEM_MESSAGES.get(role, "You are a helpful assistant.")

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("human",  "Please help me understand: {topic}")
    ])

    # Format the messages with the provided topic
    return chat_template.format_messages(topic=topic)

# ── Test with provided example ──
role  = "teacher"
topic = "Neural Networks"

messages = build_chat_prompt(role, topic)

print("=== Task 4: ChatPromptTemplate System ===")
for msg in messages:
    print(f"[{msg.type.upper()}]: {msg.content}")

print("\n--- Interviewer Role ---")
for msg in build_chat_prompt("interviewer", "Deep Learning"):
    print(f"[{msg.type.upper()}]: {msg.content}")

print("\n--- Motivator Role ---")
for msg in build_chat_prompt("motivator", "Data Science"):
    print(f"[{msg.type.upper()}]: {msg.content}")
```

## Task 5: Input Validation Layer

```python
VALID_AUDIENCES = ["beginner", "intermediate", "expert"]
VALID_TONES     = ["formal", "casual", "fun"]

# Default fallback values
DEFAULT_AUDIENCE = "beginner"
DEFAULT_TONE     = "casual"

def validate_inputs(audience: str, tone: str) -> tuple:
    validated_audience = audience
    validated_tone     = tone

    if audience not in VALID_AUDIENCES:
        print(f"[WARNING] Invalid audience '{audience}'. "
              f"Allowed: {VALID_AUDIENCES}. Defaulting to '{DEFAULT_AUDIENCE}'.")
        validated_audience = DEFAULT_AUDIENCE

    if tone not in VALID_TONES:
        print(f"[WARNING] Invalid tone '{tone}'. "
              f"Allowed: {VALID_TONES}. Defaulting to '{DEFAULT_TONE}'.")
        validated_tone = DEFAULT_TONE

    return validated_audience, validated_tone

# ── Test with valid inputs ──
print("=== Task 5: Input Validation Layer ===")
print("Valid inputs:")
print(validate_inputs("beginner", "fun"))

# ── Test with invalid inputs ──
print("\nInvalid inputs:")
print(validate_inputs("student", "aggressive"))

# ── Test with mixed inputs ──
print("\nMixed inputs (one invalid):")
print(validate_inputs("expert", "slang"))
```

## Task 6: Prompt Generator App

```python
STYLE_TEMPLATES = {
    "teaching": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} teaching style"
    ),
    "interview": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Ask 3 {tone} interview questions about {topic} for {audience}"
    ),
    "storytelling": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} storytelling style"
    ),
}

DEFAULT_STYLE = "teaching"

def generate_prompt(topic: str, audience: str, tone: str, style: str) -> str:
   # Step 1: Validate audience and tone
    audience, tone = validate_inputs(audience, tone)

    # Step 2: Validate style
    if style not in STYLE_TEMPLATES:
        print(f"[WARNING] Invalid style '{style}'. "
              f"Allowed: {list(STYLE_TEMPLATES.keys())}. Defaulting to '{DEFAULT_STYLE}'.")
        style = DEFAULT_STYLE

    # Step 3: Select template and format
    selected_template = STYLE_TEMPLATES[style]
    return selected_template.format(topic=topic, audience=audience, tone=tone)

# ── Test cases ──
print("=== Task 6: Prompt Generator App ===")

print("\n-- Example from assignment --")
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))

print("\n-- Teaching style --")
print(generate_prompt("Python", "intermediate", "casual", "teaching"))

print("\n-- Interview style --")
print(generate_prompt("Deep Learning", "expert", "formal", "interview"))

print("\n-- Invalid inputs (triggers validation) --")
print(generate_prompt("AI", "student", "slang", "quiz"))
```

## Task 7: Template Reusability Test

```python
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# Five diverse input sets to prove reusability
input_sets = [
    {"topic": "Machine Learning",   "audience": "beginner",     "tone": "friendly"},
    {"topic": "Data Structures",    "audience": "intermediate",  "tone": "formal"},
    {"topic": "Computer Vision",    "audience": "expert",        "tone": "technical"},
    {"topic": "Natural Language Processing", "audience": "kids", "tone": "fun"},
    {"topic": "Reinforcement Learning", "audience": "students",  "tone": "casual"},
]

print("=== Task 7: Template Reusability Test ===")
print("Using ONE template with 5 different inputs:\n")

for i, inputs in enumerate(input_sets, start=1):
    # Same template, different values
    output = reusable_template.format(**inputs)
    print(f"Input {i}: {output}")
```

## Full Pipeline Demo
**User Input → Validation → Prompt Template → Dynamic Prompt Generation → Output**

```python
def run_pipeline(topic: str, audience: str, tone: str, style: str) -> str:

    print(f"\n{'='*55}")
    print(f"  INPUT  → topic='{topic}', audience='{audience}', tone='{tone}', style='{style}'")
    validated_audience, validated_tone = validate_inputs(audience, tone)
    final_prompt = generate_prompt(topic, validated_audience, validated_tone, style)
    print(f"  OUTPUT → {final_prompt}")
    print(f"{'='*55}")
    return final_prompt

print("=== FULL PIPELINE DEMO ===")
run_pipeline("Neural Networks",  "beginner",     "fun",     "storytelling")
run_pipeline("Python",           "intermediate", "casual",  "teaching")
run_pipeline("Deep Learning",    "expert",       "formal",  "interview")
run_pipeline("AI Ethics",        "unknown",      "invalid", "teaching")  # triggers validation
```
